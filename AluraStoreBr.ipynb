### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

print("Faturamento Loja 1:", faturamento_loja1)
print("Faturamento Loja 2:", faturamento_loja2)
print("Faturamento Loja 3:", faturamento_loja3)
print("Faturamento Loja 4:", faturamento_loja4)
```

# 2. Vendas por Categoria

# **6. Analise Grafica**

```python
import matplotlib.pyplot as plt
import pandas as pd

lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamentos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title("Faturamento Total por Loja", fontsize=15)
plt.xlabel("Lojas", fontsize=12)
plt.ylabel("Faturamento", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("\n\n" + "-"*50 + "\n\n")

todas_lojas = pd.concat([loja, loja2, loja3, loja4])
vendas_por_categoria = todas_lojas["Categoria do Produto"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(vendas_por_categoria, labels=vendas_por_categoria.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribuição de Vendas por Categoria de Produto", fontsize=15)
plt.axis('equal')
plt.show()

print("\n\n" + "-"*50 + "\n\n")

avaliacoes = [avaliacoes_loja1, avaliacoes_loja2, avaliacoes_loja3, avaliacoes_loja4]

plt.figure(figsize=(8, 5))
plt.bar(lojas, avaliacoes, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title("Média de Avaliações dos Clientes por Loja", fontsize=15)
plt.xlabel("Lojas", fontsize=12)
plt.ylabel("Média de Avaliação", fontsize=12)
plt.ylim(3.5, 5.0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
